"Guajira Guantananamera"

En el pintoresco pueblo costero de Caimanera ubicado en la provincia de Guantánamo , en el extremo más oriental de Cuba,  vive Carmen,  una muchacha de 18 años, que a pesar de estar tan ligada a ese pueblo y de que sus raíces esten allí y más que un simple pueblo , es su hogar. Su mayor deseo es emigrar. A medida que han pasado los años,  ella cada vez tiene más claro que su futuro no lo quiere en Caimanera.  Su familia vive de la agricultura y pues a ella no le llama la atención. Ella esta en busca de ampliar sus horizontes y de irse  para un lugar donde tenga mejor calidad de vida ,  mejores oportunidades de estudio y tenga más libertad. 

¿Por qué Carmen desea emigrar?

En este municipio esta ubicada la  Base Naval de Guantánamo y para acceder a el se necesita un permiso y debes llenar una planilla con tus datos, aunque residas ahí y tanto a Carmen como a muchas otras personas eso es muy incómodo. Para ella todo sería mucho mejor si no tuviera que pasar por esto cada día. Además de las pocas oportunidades de estudio que hay en el municipio,  no cuenta con instituciones destinadas a la Enseñanza Superior ,pues solamente cuenta con 10 escuelas .

¿Carmen será la única que se siente incorforme con las condiciones de vida en su municipio y desea emigrar?

Pues a lo largo de los años muchas personas han decidido tomar esta decisión, ya sea, hacia otro municipio, fuera de su provincia o incluso fuera del país. Desde 1989, Guantánamo presenta cada año la peor tasa migratoria interna del país. Casi todos sus municipios presentan un saldo migratorio negativo. Son muy pocas las personas que emigran hacia esta provincia.

¿ Por qué Guantánamo es el municipio más poblado?.¿Que hace que las personas prefieran emigrar hacia ese municipio ?.¿Será una buena opción para Carmen emigrar hacia la cabecera de la provincia?

Guantánamo es el municipio cabecera de la provincia .Este tiene un  nivel de desarrollo económico ,si lo comparamos con el de los demás municipios , pues en él se encuentran ubicadas grandes de las industrias de la provincia  , entre las más importantes están las fábricas de válvulas y bombas,muebles,producciones textiles (AMBAR),  en la industria alimentaria destacan la  Fábrica de conservas de frutas y vegetales Guaso,el Combinado Lácteo Anacaona,  así como el Combinado Cárnico.

La base ecónomica fundamental del municipio es agropecuaria e inndustrial. Pero a pesar de poseer un mayor desarrollo ecónomico, el salario medio estimados del municipio durante el período de 2017-2022 osciló entre 554 y 3874 pesos. Pero a pesar de que este es un factor muy importante a la hora de analizar el porque de la emigracíon,pues mayormente las personas emigran en busca de mejores oportunidades de empleo y de salarios. Pero en el caso específico de este municipio. ¿Por qué será?

```python
# Gráfico de líneas

#importar las bibliotecas
import pandas as pd 
import matplotlib.pyplot as plt

#Establecer mis datos
x=[2017, 2018,2019,2020,2021,2022]
y1=[760,719,989,1580,5055,5022]
y2=[554,655,782,1133,3791,3874]

#graficar las líneas
fig,ax=plt.subplots()
ax.plot(x,y1,label="Caimanera")
ax.plot(x,y2,label="Guantánamo")
ax.legend(["Caimanera","Guantánamo"])
 
#Título del gráfico 
plt.title("Salario Medio")
plt.show()#Mostrar el gráfico






```

Pues mayormente las personas que viven en zonas de montañas,  en lugares de difícil acceso, le es muy complicado acceder a distintos servicios ,  como son la salud ,  la educación y el servicio de electrificación y toman la decisión de emigrar hacia el municipio capital , para facilitarse un poco más la vida y mejorar , dado q en la cabecera se encuentran gran parte de los hospitales de la provincia y en el sector educacional , se ubican un total de 90 escuelas, entre ellas están la Universidad de Guantánamo y Facultad de Ciencias Médicas destinadas a la educación superior.


```python
#Importar las bibliotecas
import pandas as pd
import matplotlib.pyplot as plt


#Datos
y=["Total","Primaria","Media","Especial","Adultos","Superior"]
x1=[90,43,24,10,6,2]#altura de las barras para Guantanamo
x2=[10,6,2,1,1,0]#altura de las barras para Caimanera




fig,ax=plt.subplots()

#Crear las barras
bars1=ax.barh(y,x1,height=0.30,label="Guantánamo",color="red")
bars2=ax.barh(y,x2,left=x1,height=0.30,label="Caimanera",color="orange")

#Etiquetas,titulo,leyenda
ax.set_xlabel('Escuelas')
ax.set_title(' Situación Educacional')

plt.gca()
ax.legend()
plt.show()
```

¿Se verá afectada la economía por la disminución considerable de la población en la provincia,teniendo en cuenta la decadente situación con la migración?

Uno de los sectores que se ve más afectado debido a la disminución de la población es el sector de la economía principalmente en la agricultura, teniendo en cuenta que mayormente las personas emigran hacia las zonas urbanas, lo que conlleva a la reducción de la fuerza productiva , viéndose reflejado  en la disminución de la producción.

En el 2013 la provincia produjo un total de 62740.6 toneladas de viandas mientras que en el 2022 ,fueron 40936, por otro lado en la producción de hortalizas también se vio una gran disminución ,en el 2013 se produjo 36887.3 toneladas mientras que en 2022 se produjo 10827.5 toneladas.







¿Las personas desearían emigrar hacia Guantánamo?.¿Cuáles son los principales motivos por el que alguien tomaría la decisión de emigrar?

Para darles respuestas a estas interrogantes se hizo una encuesta ,en la que participaron 36 personas ,de ellas solamente 3 eran de Guantánamo.Los resultados de la encuesta arrojaron que ninguno de los encuestados quiere emigrar hacia la provincia ,y entre los principales destinos para emigrar se encuentran La Habana con el 18.2% y Matanzas con el 6.1%.Entre los principales motivos por los cuales las personas emigrarían ,con el 69.2% esta el que la carrera que escogieron se estudia en otra provincia,mientras que el 15,4% lo haría porque en otra provincia tuviera mejores oportunidades de empleo.
